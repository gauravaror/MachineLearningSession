```python
reg = LinearRegression()
```

```python
labels = data['price']
conv_dates = [1 if values == 2014 else 0 for values in data.date ]
data['date'] = conv_dates
train1 = data.drop(['id', 'price'],axis=1)
```

### Test train split

The machine is the student and train data is the syllabus and test data is the exam. we see how much the machine has scored and if it scores well are model is successful.

We would distribute our collection into two parts. On a part of data we will train our system and on another part we will test the accuracy of our system

```python
from sklearn.cross_validation import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(train1 , labels , test_size = 0.10,random_state =2)
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

Our linear regression model scored 73% accuracy on our test data set.

But that's much lower than what we hope to achieve. We can use a powerfule algorithm.
Let's try running our algorithm on Gradient Boosting Regression

```python
from sklearn import ensemble
clf = ensemble.GradientBoostingRegressor(n_estimators = 400, max_depth = 5, min_samples_split = 2,
          learning_rate = 0.1, loss = 'ls')
```

```python
clf.fit(x_train, y_train)
```

```python
clf.score(x_test,y_test)
```

```python
t_sc = np.zeros(400,dtype=np.float64)
```

```python
y_pred = reg.predict(x_test)
```

```python
for i,y_pred in enumerate(clf.staged_predict(x_test)):
    t_sc[i]=clf.loss_(y_test,y_pred)
```

```python
testsc = np.arange(400)+1
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(testsc,clf.train_score_,'b-',label= 'Set dev train')
plt.plot(testsc,t_sc,'r-',label = 'set dev test')
```
